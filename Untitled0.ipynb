```python
fotografia = cv2.imread('foto.jpeg')
plt.imshow(fotografia)
```

```python
# Canais divididos 
b,g,r = cv2.split(fotografia)
```

```python
# Canal R (red, vermelho) em tons de cinza 
plt.imshow(r, cmap= 'gray')
```

```python
# Canal R (vermelho) 
plt.imshow( r, cmap = None)
```
